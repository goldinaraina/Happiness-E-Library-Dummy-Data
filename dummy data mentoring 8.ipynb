```python
import csv
from datetime import timedelta
import random
from faker import Faker

fake = Faker()

def generate_libraries_data(num_records):
    # menentukan path file csv
    file_path = r'C:\Users\Lenovo\Downloads\mentoring_8_goldina\libraries.csv'
    # membuka file csv untuk ditulis
    with open(file_path, 'w', newline='') as csvfile:
        # menentukan nama kolom
        fieldnames = ['library_id', 'name', 'address', 'established_date']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        # mengisi file dengan data library palsu
        for i in range(1, num_records + 1):
            writer.writerow({
                'library_id': i,
                'name': fake.company(),
                'address': fake.address(),
                'established_date': fake.date_between(start_date='-50y', end_date='-1y')
            })

def generate_books_data(num_records):
    file_path = r'C:\Users\Lenovo\Downloads\mentoring_8_goldina\books.csv'
    with open(file_path, 'w', newline='') as csvfile:
        fieldnames = ['book_id', 'library_id', 'title', 'author', 'genre', 'publisher', 'publication_date', 'language', 'isbn', 'quantity']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for i in range(1, num_records + 1):
            writer.writerow({
                'book_id': i,
                'library_id': random.randint(1, 5),  # random angka library_id dari 1 sampai 5
                'title': fake.catch_phrase(),
                'author': fake.name(),
                'genre': fake.word(),
                'publisher': fake.company(),
                'publication_date': fake.date_between(start_date='-10y', end_date='today'),
                'language': fake.language_code(),
                'isbn': fake.isbn13(),
                'quantity': random.randint(1, 2)
            })

def generate_users_data(num_records):
    file_path = r'C:\Users\Lenovo\Downloads\mentoring_8_goldina\users.csv'
    with open(file_path, 'w', newline='') as csvfile:
        fieldnames = ['user_id', 'username', 'email', 'password', 'registration_date']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for i in range(1, num_records + 1):
            writer.writerow({
                'user_id': i,
                'username': fake.user_name(),
                'email': fake.email(),
                'password': fake.password(),
                'registration_date': fake.date_between(start_date='-5y', end_date='today')
            })

def generate_loans_data(num_records):
    file_path = r'C:\Users\Lenovo\Downloads\mentoring_8_goldina\loans.csv'
    user_loan_counts = {i: 0 for i in range(1, 51)}  # dictionary untuk melacak jumlah peminjaman per user
    with open(file_path, 'w', newline='') as csvfile:
        fieldnames = ['loan_id', 'book_id', 'user_id', 'loan_date', 'due_date', 'return_date']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        loan_id = 1
        attempts = 0  # inisialisasi counter attempt
        while loan_id <= num_records and attempts < 1000:  # batas attempt untuk mencegah infinite loop
            loan_date = fake.date_between(start_date='-1y', end_date='today')
            due_date = loan_date + timedelta(days=14)
            return_date = loan_date + timedelta(days=random.randint(1, 14)) if random.random() < 0.8 else None
            
            user_id = random.randint(1, 50)  # random user_id dari 1 sampai 50
            if user_loan_counts[user_id] >= 2:
                attempts += 1  # tambah attempt jika user sudah meminjam 2 buku
                continue  # lewati iterasi ini jika user sudah meminjam 2 buku
            
            book_id = random.randint(1, 100)  # random book_id dari 1 sampai 100
            
            user_loan_counts[user_id] += 1  # tambah jumlah peminjaman user

            writer.writerow({
                'loan_id': loan_id,
                'book_id': book_id,
                'user_id': user_id,
                'loan_date': loan_date,
                'due_date': due_date,
                'return_date': return_date
            })
            loan_id += 1
            attempts = 0  # reset attempt setelah berhasil menulis data

def generate_holds_data(num_records):
    file_path = r'C:\Users\Lenovo\Downloads\mentoring_8_goldina\holds.csv'
    with open(file_path, 'w', newline='') as csvfile:
        fieldnames = ['hold_id', 'book_id', 'user_id', 'hold_date', 'expiry_date', 'status']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for i in range(1, num_records + 1):
            book_id = random.randint(1, 100)  # random book_id dari 1 sampai 100
            user_id = random.randint(1, 50)   # random user_id dari 1 sampai 50
            hold_date = fake.date_between(start_date='today', end_date='+1y')
            expiry_date = hold_date + timedelta(days=random.randint(7, 30))
            writer.writerow({
                'hold_id': i,
                'book_id': book_id,
                'user_id': user_id,
                'hold_date': hold_date,
                'expiry_date': expiry_date,
                'status': random.choice(['active', 'released', 'borrowed'])
            })

# contoh penggunaan:
generate_libraries_data(5)   # generate 5 data libraries
generate_books_data(100)     # generate 100 data books
generate_users_data(50)      # generate 50 data users
generate_loans_data(200)     # generate 200 data loans
generate_holds_data(100)     # generate 100 data holds
```
